```python
print("Rock Paper Scissors Game")

user = input("Choose rock, paper, or scissors: ")
computer = "rock"

if user == computer:
    print("It's a tie!")
elif (user == "paper" and computer == "rock") or \
     (user == "rock" and computer == "scissors") or \
     (user == "scissors" and computer == "paper"):
    print("You win!")
else:
    print("Computer wins!")
```
